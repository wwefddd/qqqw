단계2) 교재 77쪽의 프로그램 상에서 휴리스틱 함수 h1을 다음과 같이 h2로 수정하여 프로그램 실행

    > h1(n) = 목표상태와 비교하여 잘못 놓여진 타일의 개수

    > h2(n) = 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리의 합

```python
import queue

# 상태를 나타내는 클래스, f(n) 값을 저장한다.
class State:
  def __init__(self, board, goal, depth=0):
    self.board = board			# 현재의 보드 상태
    self.depth = depth			# 깊이
    self.goal = goal			# 목표 상태

  # i1과 i2를 교환하여서 새로운 상태를 반환한다.
  def get_new_board(self, i1, i2, depth):
    new_board = self.board[:]
    new_board[i1], new_board[i2] = new_board[i2], new_board[i1]
    return State(new_board, self.goal, depth)

  # 자식 노드를 확장하여서 리스트에 저장하여서 반환한다.
  def expand(self, moves):
    result = []
    i = self.board.index(0)		# 숫자 0(빈칸)의 위치를 찾는다.
    if not i in [0, 3, 6] :		# LEFT 연산자
      result.append(self.get_new_board(i, i-1, moves))
    if not i in [0, 1, 2] :		# UP 연산자
      result.append(self.get_new_board(i, i-3, moves))
    if not i in [2, 5, 8]:		# RIGHT 연산자
      result.append(self.get_new_board(i, i+1, moves))
    if not i in [6, 7, 8]:		# DOWN 연산자
      result.append(self.get_new_board(i, i+3, moves))
    return result

  # f(n)을 계산하여 반환한다.
  def f(self):
    return self.h()+self.g()

  # 휴리스틱 함수 값인 h(n)을 계산하여 반환한다.
  # 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리의 합을 반환한다. (즉, h(n) = h2(n))
  def h(self):
      score = 0
      # 시간복잡도 : O(타일 리스트 크기)
      # d[목표 값] = 목표값 인덱스
      d = [0 for _ in range(len(self.board))]
      for i in range(len(self.goal)):
        d[self.goal[i]] = i

      # 실제 거리 찾기
      for i in range(len(self.board)):
        element = self.board[i]
        if element == 0 : continue;
        goal_index = d[element]
        # 1차원 array index를 2차원 3x3 행렬에 mapping
        boardi = i // 3
        boardj = i % 3
        goali = goal_index // 3
        goalj = goal_index % 3
        # 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리를 더한다
        score += abs(boardi - goali) + abs(boardj - goalj)
      return score


      # 시간복잡도 : O(타일 리스트 크기**2)
      # def h2(self):
      #     score = 0
      #     for i in range(1, 9):
      #         boardidx = self.board.index(i)
      #         goalidx = self.goal.index(i)

      #         # 1차원 array index를 2차원 3x3 행렬에 mapping
      #         boardi = boardidx // 3
      #         boardj = boardidx % 3
      #         goali = goalidx // 3
      #         goalj = goalidx % 3

      #         # 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리를 더한다
      #         score += abs(boardi - goali) + abs(boardj - goalj)
      #     return score


  # 시작 노드로부터의 깊이를 반환한다.
  def g(self):
    return self.depth

  def __eq__(self, other):
    return self.board == other.board

  def __ne__(self, other):
    return self.board != other.board

  # 상태와 상태를 비교하기 위하여 less than 연산자를 정의한다.
  def __lt__(self, other):
    return self.f() < other.f()

  def __gt__(self, other):
    return self.f() > other.f()

  # 객체를 출력할 때 사용한다.
  def __str__(self):
    return f"f(n)={self.f()} h(n)={self.h()} g(n)={self.g()}\n"+\
    str(self.board[:3]) +"\n"+\
    str(self.board[3:6]) +"\n"+\
    str(self.board[6:]) +"\n"

# 초기 상태
puzzle = [2, 8, 3,
          1, 6, 4,
          7, 0, 5]
# 목표 상태
goal = [1, 2, 3,
        8, 0, 4,
        7, 6, 5]

# open 리스트는 우선순위 큐로 생성한다.
open_queue = queue.PriorityQueue()
open_queue.put(State(puzzle, goal))

closed_queue = [ ]
depth = 0
count = 0

while not open_queue.empty():
  current = open_queue.get()
  count += 1
  print(count)
  print(current)
  if current.board == goal:
      print("탐색 성공")
      break
  depth = current.depth+1
  for state in current.expand(depth):
    if state not in closed_queue and state not in open_queue.queue :
      open_queue.put(state)
  closed_queue.append(current)
else:
  print ('탐색 실패')
```

(단계3) h1과 h2 각각에 대해 탐색 알고리즘의 효율성을 비교

    > 랜덤하게 설정된 10가지의 서로다른 초기상태 각각에 대해 h1을 사용한 탐색비용과 h2를 사용한 탐색 비용을 비교

    > 탐색비용은 탐색과정에서 생성된 모든 노드의 개수

```python
import queue
import random

from wcwidth import wcswidth  # 한글이 포함된 문자열에 간격 맞추기 솔루션을 제공하는 라이브러리

def fmt(x, w, align='r'): # align의 기본값은 'r : right' 출처 : https://fhaktj8-18.tistory.com/entry/str-interval-setting
    """ 동아시아문자 폭을 고려하여, 문자열 포매팅을 해 주는 함수.
    w 는 해당 문자열과 스페이스문자가 차지하는 너비.
    align 은 문자열의 수평방향 정렬 좌/우/중간.
    """
    x = str(x) # 해당 문자열
    l = wcswidth(x) # 문자열이 몇자리를 차지하는지를 계산.
    s = w-l # 남은 너비 = 사용자가 지정한 전체 너비 - 문자열이 차지하는 너비
    if s <= 0:
        return x
    if align == 'l':
        return x + ' '*s
    if align == 'c':
        sl = s//2 # 변수 좌측
        sr = s - sl # 변수 우측
        return ' '*sl + x + ' '*sr
    return ' '*s + x

# 상태를 나타내는 클래스, f(n) 값을 저장한다.
class State:
  def __init__(self, board, goal, hnum, depth=0):
    self.hnum = hnum        # 휴리스틱 함수 h1(n), h2(n) 중 어떤걸 택할지 정하는 변수
    self.board = board			# 현재의 보드 상태
    self.depth = depth			# 깊이
    self.goal = goal			# 목표 상태



  # i1과 i2를 교환하여서 새로운 상태를 반환한다.
  def get_new_board(self, i1, i2, depth):
    new_board = self.board[:]
    new_board[i1], new_board[i2] = new_board[i2], new_board[i1]
    return State(new_board, self.goal, self.hnum, depth)

  # 자식 노드를 확장하여서 리스트에 저장하여서 반환한다.
  def expand(self, moves):
    result = []
    i = self.board.index(0)		# 숫자 0(빈칸)의 위치를 찾는다.
    if not i in [0, 3, 6] :		# LEFT 연산자
      result.append(self.get_new_board(i, i-1, moves))
    if not i in [0, 1, 2] :		# UP 연산자
      result.append(self.get_new_board(i, i-3, moves))
    if not i in [2, 5, 8]:		# RIGHT 연산자
      result.append(self.get_new_board(i, i+1, moves))
    if not i in [6, 7, 8]:		# DOWN 연산자
      result.append(self.get_new_board(i, i+3, moves))
    return result

  # hnum에 따라서 h1과 h2 중 하나를 선택하고, f(n)을 계산하여 반환한다.
  def f(self):
      if self.hnum == 1 :
          return self.h1()+self.g()
      else :
          return self.h2()+self.g()

  # 휴리스틱 함수 값인 h(n)을 계산하여 반환한다.
  # h1(n) = 현재 제 위치에 있지 않은 타일의 개수를 계산하여 반환한다.
  def h1(self):
    score = 0
    for i in range(9):
        if self.board[i]!=0 and self.board[i] != self.goal[i]:
            score += 1
    return score

  # h2(n) = 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리의 합을 반환한다.
  def h2(self):
      score = 0
      # 시간복잡도 : O(타일 리스트 크기)
      # d[목표 값] = 목표값 인덱스
      d = [0 for _ in range(len(self.board))]
      for i in range(len(self.goal)):
        d[self.goal[i]] = i

      # 실제 거리 찾기
      for i in range(len(self.board)):
        element = self.board[i]
        if element == 0 : continue;
        goal_index = d[element]
        # 1차원 array index를 2차원 3x3 행렬에 mapping
        boardi = i // 3
        boardj = i % 3
        goali = goal_index // 3
        goalj = goal_index % 3
        # 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리를 더한다
        score += abs(boardi - goali) + abs(boardj - goalj)
      return score


      # 시간복잡도 : O(타일 리스트 크기**2)
      # def h2(self):
      #     score = 0
      #     for i in range(1, 9):
      #         boardidx = self.board.index(i)
      #         goalidx = self.goal.index(i)

      #         # 1차원 array index를 2차원 3x3 행렬에 mapping
      #         boardi = boardidx // 3
      #         boardj = boardidx % 3
      #         goali = goalidx // 3
      #         goalj = goalidx % 3

      #         # 각 타일의 현재위치와 목표위치 사이의 맨하탄 거리를 더한다
      #         score += abs(boardi - goali) + abs(boardj - goalj)
      #     return score

  # 시작 노드로부터의 깊이를 반환한다.
  def g(self):
    return self.depth

  def __eq__(self, other):
    return self.board == other.board

  def __ne__(self, other):
    return self.board != other.board

  # 상태와 상태를 비교하기 위하여 less than 연산자를 정의한다.
  def __lt__(self, other):
    return self.f() < other.f()

  def __gt__(self, other):
    return self.f() > other.f()

  # 객체를 출력할 때 사용한다.
  def __str__(self):
      if self.hnum == 1 :
        return f"f(n)={self.f()} h1(n)={self.h2()} g(n)={self.g()}\n"+\
        str(self.board[:3]) +"\n"+\
        str(self.board[3:6]) +"\n"+\
        str(self.board[6:]) +"\n"
      else :
        return f"f(n)={self.f()} h2(n)={self.h2()} g(n)={self.g()}\n"+\
        str(self.board[:3]) +"\n"+\
        str(self.board[3:6]) +"\n"+\
        str(self.board[6:]) +"\n"

# 리스트의 반전(inversion) 개수가 홀수인지 짝수인지 판단하는 함수이다.
def inversions(mylist):
    inversion = 1
    testlist = mylist[:]
    testlist.remove(0) # 공백은 제거
    for i in range(len(testlist) - 1):
        for j in range(i,len(testlist)):
            if testlist[i]  > testlist[j] :
                inversion *= -1 # -1 이면 inversion 홀수개 1 이면 짝수개
    return inversion

# h1(n)을 사용했을때의 비용과 h2(n)을 사용했을때의 비용에 따라서 효율성을 비교하는 함수이다.
def efficiency(h1cost, h2cost):
    if h1cost < h2cost :
        return "h1 > h2"
    elif h1cost > h2cost :
        return "h1 < h2"
    else :
        return "h1 = h2"

# h1 탐색비용 리스트
h1costlist = []

# h2 탐색비용 리스트
h2costlist = []

# 초기 상태 리스트
puzzlelist = []

# 효율 리스트
efficiencylist = []

# 목표 상태
goal = [1, 2, 3,
        8, 0, 4,
        7, 6, 5]

howmanypuzzle = int(input("몇개의 초기상태를 생성하겠습니까? : ")) # 몇개의 초기상태를 사용할건지 묻는다.

while len(puzzlelist) < howmanypuzzle : # 총 howmanypuzzle개의 초기상태를 생성한다.


    puzzle = [1,2,3,
              4,5,6,
              7,8,0]

    random.shuffle(puzzle) # 랜덤한 초기상태를 생성한다.

    if inversions(puzzle) == inversions(goal) : # 초기상태와 목표상태의 inversion 홀짝성이 같으면 (즉, 해결 할 수 있는 초기상태이면) puzzlelist에 추가

        puzzlelist.append(puzzle)

    else :
        # print("해결 가능한 상태가 아닙니다. ")
        continue

for puzzle in puzzlelist: # 퍼즐리스트에 모은 랜덤 초기상태들마다 실행
    for hnum in [1,2]: # h(n)을 h1(n)을 사용할때와 h2(n)을 사용할때 따로 실행
        # open 리스트는 우선순위 큐로 생성한다.
        open_queue = queue.PriorityQueue()
        open_queue.put(State(puzzle, goal, hnum))

        closed_queue = [ ]
        depth = 0
        count = 0

        while not open_queue.empty():
          current = open_queue.get()
          count += 1
          print(count)
          print(current)
          if current.board == goal:
              print("탐색 성공")
              break
          depth = current.depth+1
          for state in current.expand(depth):
            if state not in closed_queue and state not in open_queue.queue :
              open_queue.put(state)
          closed_queue.append(current)
        else:
          print ('탐색 실패')


        # 탐색 비용 = Open_queue 원소 개수 + Closed_queue 원소 개수 + 1
        cost = open_queue.qsize() + len(closed_queue) + 1
        # h1(n)을 사용할때 h1비용을 구해 h1costlist에 넣는다
        if hnum == 1 :
            h1cost = cost
            print(f" h1 비용 : {h1cost}")
            h1costlist.append(h1cost)
        # h2(n)을 사용할때 h2비용을 구해 h2costlist에 넣는다
        else :
            h2cost = cost
            print(f" h2 비용 : {h2cost}")
            h2costlist.append(h2cost)
    # h1 비용과 h2 비용을 비교해 어떤 함수가 더 효율적인지 구한걸 efficiencylist에 넣는다.
    efficiencylist.append(efficiency(h1cost,h2cost))
    print(f" h1 비용 : {h1cost}, h2 비용 : {h2cost}")



print(" ")
print("---------------------------------------------------------")
print(" ")
print("초기 상태                    h1 비용         h2 비용         효율성")
# 문자열 간격 맞춰서 출력. 출처 : https://fhaktj8-18.tistory.com/entry/str-interval-setting
for pz, h1, h2, ef in zip(puzzlelist, h1costlist, h2costlist, efficiencylist) :
    print('{} {} {} {}'.format(fmt(str(pz),18,'l'), fmt(str(h1),15,'l'), fmt(str(h2), 15,'l'), fmt(str(ef), 15,'l')))
#for i in range(10) : print(f"{puzzlelist[i]} {h1costlist[i]} {h2costlist[i]} {efficiencylist[i]}")
```
