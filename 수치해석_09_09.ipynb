```python
import numpy as np
import matplotlib.pyplot as plt
#taylor series exp
N = 101 # 점 개수
x = np.linspace(-1, 1, N) # N 개의 점으로 [-1,1] 을 동일한 간격으로 분할
h = 2 / ( N - 1) # 그 동일한 간격의 길이 interval [a,b] 이면 간격은 (b-a)/(N-1)
fx = np.exp(x)

# 테일러 다항식
p1 = 1 + x
p2 = 1 + x + 0.5 * x ** 2

# Visualization
plt.plot(x, fx, '-')
plt.plot(x, p1, 'r-')
plt.plot(x, p2, 'g-')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
# f = exp, a = 0
N = 101 # 점 개수
x = np.linspace(-1, 1, N) # N 개의 점으로 [-1,1] 을 동일한 간격으로 분할
h = 2 / ( N - 1) # 그 동일한 간격의 길이 interval [a,b] 이면 간격은 (b-a)/(N-1)
fx = np.exp(x)

# n차 테일러 다항식
deg = 2
pn = np.zeros(N)
for j in range(0, deg+1):
  pn[:] = pn[:] + 1/math.factorial(j) * x**j

# Visualization
plt.plot(x, fx, '-')
plt.plot(x, pn, 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
