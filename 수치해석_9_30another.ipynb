```python
# 200677 수학과 김준성 수치해석및실습 5주차 과제
# Bisection method for rootfinding using a function f(x)
# Ex. Try f (x) := x**2 - 1 = 0

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  fx = x**2 - 1
  return fx

def main():

  a, b = 0.7, 2.718
  etol = 1.0e-7
  max_iter = 100

  i = 1

  while i < max_iter:

    c = (a + b ) / 2

    fb = f(b)
    fc = f(c)

    if b-c <= etol:
      print("c = %.16f is our numerical solution!" % c)
      print("|f(c)| is %.16f." % np.abs ( fc ))
      break
    elif np.sign(fb) * np.sign(fc) <= 0:
      a = c
    else :
      b = c
    print("%d iteration, c = %.16f, |f(c)| = %.16f " % (i, c, np.abs(fc)))
    i = i + 1

main()
```

```python

# Newton’s (or Newton-Raphson) method for rootfinding
# Ex. Define f (x) and main function.
# Ex. Try f (x) := x**2 - 1 = 0

import numpy as np
import matplotlib.pyplot as plt


def f(x):
  fx = x**2 - 1
  return fx

def dfd(x):
  dfdx = 2 * x
  return dfdx

def main():
  x_n = 1.5
  etol = 1.0e-7
  max_iter = 100
  alpha = 1

  print("n,   x_n,      f(x_n),  x_{n} - x_{n-1}, alpha - x_{n-1}")
  print("%d, %.8f, %.2e" %(0, x_n, x_n**6 -x_n - 1))

  n = 1
  while n < max_iter:

    fx = f(x_n)
    dfdx = dfd(x_n)

    temp = x_n
    x_n = x_n - fx/dfdx
    fx = f(x_n)

    print("%d, %.8f, %.2e, %.2e, %.2e" % (n, x_n, fx, x_n-temp, alpha-temp))

    if np.abs(x_n - temp) <= etol :
      break

    n = n+1

main()
```

```python


# Secant method for rootfinding
# Ex. Define f (x) and main function.
# Ex. Try f (x) := x**2 - 1 = 0

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  fx = x**2 - 1
  return fx

def main():

  x0 = 0.7
  x1 = 1.89
  etol = 1.0e-7
  max_iter = 100
  alpha = 1

  print("n,   x_n,      f(x_n),  x_{n} - x_{n-1}, alpha - x_{n-1}")
  print("%d, %.8f, %.2e" %(0, x0, x0**6 -x0 - 1))
  print("%d, %.8f, %.2e, %.2e" %(1, x1, x1**6 -x1 - 1, x1-x0))

  n = 2
  while n < max_iter:

    fx0 = f(x0)
    fx1 = f(x1)

    temp = x1
    x1 = x1 - fx1 * (x1 - x0 ) / (fx1 - fx0)
    x0 = temp
    fx1 = f(x1)

    print("%d, %.8f, %+.2e, %+.2e, %+.2e" % (n, x1, fx1, x1-x0, alpha-x0))

    if np.abs(x1 - x0) <= etol :
      break

    n = n+1

main()
```
