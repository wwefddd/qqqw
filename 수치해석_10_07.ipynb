Fixed point iteration I1.

```python
import numpy as np
import matplotlib.pyplot as plt

x_n = 2.5
etol = 1.0e-7
max_iter = 10000
alpha = np.sqrt(5)

print("n,     x_n, f(x_n), x_{n}, x_{n-1}, alpha - x_{n-1}")
print("%d, %.8f, %.2e" % (0, x_n, x_n**2 - 5))

n = 1
while n < max_iter:

  # x_n을 x_n+1 로 업데이트 하는 과정
  temp = x_n
  # 아래 부분이 x_n 과 x_n+1에 관련된 점화식. I2, I3, I4 코드를 짤 때에도 템플릿을 그대로 갔다 쓰고 아래 부분의 점화식만 바꾸면 된다.
  x_n = 5 + x_n - x_n**2

  fx = x_n**2 - 5

  print("%d, %.8f, %.2e, %.2e, %.2e" % (n, x_n, fx, x_n - temp, alpha - temp))

  # x_n이 수렴할려고 하면 종료 즉, |x_n+1 - x_n| < 입실론 이면 수렴한걸로 친다
  if np.abs(x_n - temp) <= etol:
    break

  n = n + 1
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_n = 2.5
etol = 1.0e-7
max_iter = 1000
alpha = np.sqrt(5)

print("n,     x_n, f(x_n), x_{n}, x_{n-1}, alpha - x_{n-1}")
print("%d, %.8f, %.2e" % (0, x_n, x_n**2 - 5))

# c 가 -1/(루트 5) 와 0 사이 음수이면 되는듯?
c = -1/1000

n = 1
while n < max_iter:


  # x_n을 x_n+1 로 업데이트 하는 과정
  temp = x_n
  # 아래 부분이 x_n 과 x_n+1에 관련된 점화식.
  x_n = x_n + c * (x_n**2 - 5)
  fx = x_n**2 - 5

  print("%d, %.8f, %.2e, %.2e, %.2e" % (n, x_n, fx, x_n - temp, alpha - temp))

  # x_n이 수렴할려고 하면 종료 즉, |x_n+1 - x_n| < 입실론 이면 수렴한걸로 친다
  if np.abs(x_n - temp) <= etol:
    break

  n = n + 1
```
