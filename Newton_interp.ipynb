```python
import numpy as np
import matplotlib.pyplot as plt

N = 7 # number of given points
M = 100 # number of sample points
xp = np.zeros(N)
yp = np.zeros(N)
Px = np.zeros(M) # Lagrange' formula
DD = np.zeros(( N-1, N-1 )) # Divided Differences
x = np.linspace(-1, N, M) # sample points (정의역 -1부터 N까지)

xp = np.linspace(0, N-1, N) # given 점 정의역 -1부터 N 까지
yp = np.random.rand(N) # 이게 fitting 할 점 (xp, yp)

# Compute 'DD'
DD[0,:] = ( yp[1:N] - yp[0:N-1] ) / ( xp[1:N] - xp[0:N-1] )
for i in range(1, N-1):
    for j in range(0, N-1-i):
        print(i,j)
        DD[i,j] = ( DD[i-1,j+1] - DD[i-1,j] ) / ( xp[j+i+1] - xp[j] )

# Compute 'P(x)'
Px[:] = yp[0] # Initialize
for i in range(0, N-1):
    temp = DD[i,0]
    for j in range(0, i+1):
        temp = ( x - xp[j] )*temp
    Px[:] = Px[:] + temp

# Visualization
plt.plot(xp, yp, 'ro')
plt.plot(x, Px, '-')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
